```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# 1. Your original data
df = pd.DataFrame(data) 

# 2. Encoding
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_array = encoder.fit_transform(df[['location']])

# 3. Convert the array back to a DataFrame with column names
# encoder.get_feature_names_out() gives names like 'location_Rasulgarh'
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(['location']))

# 4. Concatenate with the original numerical columns
# We drop 'location' because we now have the encoded version
final_df = pd.concat([df.drop('location', axis=1), encoded_df], axis=1)

# 5. Now you can check correlation!
print(final_df.corr()['price_lakh'].sort_values(ascending=False))
```

```python
# one hot encoding
data = pd.get_dummies(data, columns=['location'])
```

```python
input=data.drop(["price_lakh"],axis=1)
```

```python
output=data["price_lakh"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,output,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python

import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

# training

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python

## print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python

## on which parameters the model has been trained
regression.get_params()
```

```python

### Prediction With Test Data
reg_pred=regression.predict(x_test)
```

```python
reg_pred
```

```python
plt.scatter(y_test,reg_pred)
```

```python
## Residuals
residuals=y_test-reg_pred
residuals
```

```python

## Plot this residuals 

sns.displot(residuals,kind="kde")
```

```python

## Scatter plot with respect to prediction and residuals
## uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```
